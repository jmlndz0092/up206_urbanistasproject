It looks like the minimum household income in a census track is 9,191. (Interesting, I'd like to know what area that median household income level falls on.) While the maximum median household income in a census track is 250,001. Thats a huge gap! I definitely want to see where folks are making maximum 250,001. This can also help us when we bring in out grocery store data to help us visualize where more grocery stores are located. Is there a concentration in areas with lower income where it might be needed or in places of higher income levels, where people might be able to afford higher concentrations and more quality grocery stores.

```python
ax = LA_income.hist(column='Median HH Income', bins=50, grid=False, figsize=(12,8), color='#6f55e0', zorder=2, rwidth=0.9)

ax = ax[0]
for x in ax:

    # Switch off ticks
    x.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")

    # Draw horizontal axis lines
    vals = x.get_yticks()
    for tick in vals:
        x.axhline(y=tick, linestyle='dashed', alpha=0.4, color='#8a8888', zorder=1)

    # Remove title
    x.set_title("Median Household Income Distribution in Los Angeles", weight='bold', size=18)

    # Set x-axis label
    x.set_xlabel("Income (Dollars)", labelpad=20, size=12)

    # Set y-axis label
    x.set_ylabel("Frequency", labelpad=20, size=12)
```

Now, let's see what the top earnings in Los Angeles are...

```python
LA_income_sorted = LA_income.sort_values(by='Median HH Income',ascending = False)
```

```python
LA_income[LA_income['Median HH Income'] < 66000].plot(figsize=(12,10),column='Median HH Income',
                 legend=True, 
                 scheme='quantiles')
```

```python
LA_income_web_mercator = LA_income.to_crs(epsg=3857)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

LA_income_web_mercator[LA_income_web_mercator['Median HH Income'] < 66000].plot(ax=ax, alpha=0.8, column='Median HH Income',
                 legend=True, 
                 scheme='quantiles')

# turn the axis off
ax.axis('off')

# set a title
ax.set_title('Census Tracts with less than $66,000 Median Household Income',fontsize=14, weight='bold')

# add a basemap
ctx.add_basemap(ax)
```

By looking at the data, it seems like there's a high concentration of census tracts where people make less than $50,000. Further, most census tracts' median household income ranges from around 50,000 to almost 100,000. 100,000 is where we begin to see a drop in census tracts that make that amount or close to that amount. The frequency continues to drop, less and less census tracts have identified people that make close to 200,000 and the outlier seems to be census tracts where the median household income is 250,000. 

```python
LA_income[LA_income['Median HH Income'] > 66000].plot(figsize=(12,10),column='Median HH Income',
                 legend=True, 
                 scheme='quantiles')
```

```python
LA_income_web_mercator = LA_income.to_crs(epsg=3857)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

LA_income_web_mercator[LA_income_web_mercator['Median HH Income'] > 66000].plot(ax=ax, alpha=0.8, column='Median HH Income',
                 legend=True, 
                 scheme='quantiles')

# turn the axis off
ax.axis('off')

# set a title
ax.set_title('Census Tracts with more than $66,000 Median Household Income',fontsize=14, weight='bold')

# add a basemap
ctx.add_basemap(ax)
```

Census tracts with less than 66,000 median household income tend to be more concentrated within the central areas of LA County, whereas census tracts with more than 66,000 median household income tend to be concentrated in the surrounding, more suburban areas of LA County.
